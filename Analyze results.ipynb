```python
import pandas as pd
import numpy as np
import seaborn as sns
from tqdm import tqdm
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import glob
import pickle
```

```python
f_paths = glob.glob('pickles/*')
results=[]
for path in tqdm(f_paths):
    results.append(pickle.load(open(path,'rb')))
```

```python
df = pd.DataFrame(results)
#df = pd.read_csv('results_11_7.csv')
df = df[df['filter_approach']=='probability']
groups = ['df_name','number_of_estimators','max_number_of_branches','filter_approach']
df[[col for col in df.columns if 'auc' in col]+groups].groupby(groups).mean()
```
