```python
import pandas as pd
import numpy as np
from ExperimentSetting import *
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#df=pd.read_csv('results_10_6_18.csv')
df=pd.read_csv('experement_6_6_18.csv')
```

```python
groups=['df_name','number_of_estimators','max_number_of_branches']
df[[col for col in df.columns if 'accuracy' in col]+groups].groupby(groups).mean()
```

```python
l=[]
for index,row in df[df['number_of_estimators']==100].iterrows():
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'ensemble','auc':row['ensemble_auc']})
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'decision tree','auc':row['decision_tree_auc']})
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'new model','auc':row['new_model_auc']})
for df_name in ['winery','breast cancer','iris']:
    boxplot_data=pd.DataFrame(l)
    boxplot_data=boxplot_data[boxplot_data['df_name']==df_name]
    sns.boxplot(x='num_of_branches',hue='model',y='auc',data=boxplot_data)
    plt.title(df_name)
    plt.show()
    plt.clf()
```

```python
l=[]
for index,row in df[df['number_of_estimators']==100].iterrows():
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'ensemble','accuracy':row['ensemble_accuracy']})
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'decision tree','accuracy':row['decision_tree_accuracy']})
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'new model','accuracy':row['new_model_accuracy']})
for df_name in ['winery','breast cancer','iris']:
    boxplot_data=pd.DataFrame(l)
    boxplot_data=boxplot_data[boxplot_data['df_name']==df_name]
    sns.boxplot(x='num_of_branches',hue='model',y='accuracy',data=boxplot_data)
    plt.title(df_name)
    plt.show()
    plt.clf()
```

```python
l=[]
for index,row in df[df['number_of_estimators']==100].iterrows():
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'ensemble','average_depth':row['ensemble_average_depth']})
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'decision tree','average_depth':row['decision_tree_average_depth']})
    l.append({'df_name':row['df_name'],'num_of_branches':row['max_number_of_branches'],'model':'new model','average_depth':row['new_model_average_depth']})
for df_name in ['winery','breast cancer','iris']:
    boxplot_data=pd.DataFrame(l)
    boxplot_data=boxplot_data[boxplot_data['df_name']==df_name]
    boxplot_data['average_depth']=[np.log(i) for i in boxplot_data['average_depth']]
    sns.boxplot(x='num_of_branches',hue='model',y='average_depth',data=boxplot_data)
    plt.title(df_name+' (log)')
    plt.show()
    plt.clf()
```
